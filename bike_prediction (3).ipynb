```python
#Q5. data split into training and validation sets and build a simple Linear Regression model.  
```

```python
X = train_df.drop(['count','count_log'], axis = 1)
Y = train_df['count_log']
Y = np.log1p(Y)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_log_error
```

```python
X_train_df, X_val, y_train, y_val = train_test_split(X, Y, test_size= 0.2, random_state = 42)
```

```python
X_train_df.drop(columns=['date'],errors ='ignore', inplace = True)
X_val.drop(columns=['date'], errors='ignore', inplace =True)
X_train_df.dtypes
```

```python
lr = LinearRegression()
lr.fit(X_train_df, y_train)
```

```python
#Q6. Model Improvements
```

```python
#A.Polynomial Regression(degree = 2) 
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 2, include_bias=False)
X_train_poly = poly.fit_transform(X_train_df)
X_val_poly = poly.transform(X_val)
```

```python
lr_poly=LinearRegression()
lr_poly.fit(X_train_poly, y_train)
```

```python
Y_pred_poly = lr_poly.predict(X_val_poly)
poly_rmsle = np.sqrt(mean_squared_log_error(np.expm1(y_val), np.expm1(Y_pred_poly)))
poly_rmsle
```

```python
#B. Ridge Regression
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_log_error
import numpy as np
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_poly, y_train)
y_pred_ridge = ridge.predict(X_val_poly)
ridge_rmsle = np.sqrt(mean_squared_log_error(np.expm1(y_val), np.expm1(y_pred_ridge)))

print("Ridge RMSLE:", ridge_rmsle)
```

```python
#C. Lasso Regression
```

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_poly)
X_val_scaled = scaler.transform(X_val_poly)
```

```python
lasso = Lasso(alpha=0.001, max_iter=5000)
lasso.fit(X_train_scaled, y_train)
```

```python
Y_pred_lasso = lasso.predict(X_val_poly)
Y_pred_lasso = np.maximum(Y_pred_lasso, 0)
lasso_rmsle = np.sqrt(mean_squared_log_error(np.expm1(y_val),np.expm1(Y_pred_lasso)))
lasso_rmsle
```

```python
#Q7 when comparing all three models, Ridge + polynomial has best generalization. 
# whereas Lasso + polynomial is Sparse but slightly weaker 
# and Polynomial(degree 2) Captures non-linearity.
# therefore Ridge Regression with Polynomial Features is the Best Model.
```

```python
#Q8.Residual Plot(Best Model)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

residuals = y_val - y_pred_ridge

plt.figure(figsize=(6,4))
sns.scatterplot(x=y_pred_ridge, y=residuals)
plt.axhline(0, color='red')
plt.xlabel("Predicted Log Count")
plt.ylabel("Residuals")
plt.title("Residual Plot - Ridge Regression")
plt.show()
```

```python
#Q9. Ridge + polynomial is the winning model because polynomial features captures non-linearity demanded patterns and 
# Ridge  prevents co-efficient explosion.
#Handles multi-collinearity (temp Vs atemp)
#Generalizes better on unseen data
```

```python
#Q10. RMSLE penalizes under-prediction less than RMSE because RMSLE compares log values, so errors ar small values matter less.
#This is ideal for bike demand where underestimating low counts is less severe than missing peak demand.
```

```python
#Q11. The Trade-offs between Model Simplicity and Predictive Power is 
#Model simplicity has simple model, easy to explain, low overfitting and Fast whereas
#Predictive Power has complex model, better accuracy, needs regularization and computationally expensive.
#Best choice is balanced complexity(Ridge + polynomial).
```
